## Problem Description

Create a beginner-friendly Google Colab Python program using the holidays library. My birthday is February 15, 2001. Find all U.S. holidays that occurred one day before, on the day of, and one day after my birthday. Include comments explaining the code and print the results clearly.

## Methodology

The Python program used the holidays library and Python’s datetime module. The code created a date object for February 15, 2001, then used timedelta(days=1) to create the dates one day before and one day after the birthday. The program created a U.S. holiday calendar for 2001 and checked each of the three dates for an associated holiday.

```python
!pip install holidays

import holidays
from datetime import date, timedelta

birthday = date(2001, 2, 15)

us_holidays = holidays.US(years=[2001])

dates_to_check = [
    birthday - timedelta(days=1),
    birthday,
    birthday + timedelta(days=1)
]

print("U.S. Holidays Near My Birthday")
print("-" * 40)

found_holiday = False

for current_date in dates_to_check:
    holiday_name = us_holidays.get(current_date)

    if holiday_name:
        print(f"{current_date}: {holiday_name}")
        found_holiday = True
    else:
        print(f"{current_date}: No U.S. holiday found")

if not found_holiday:
    print("\nNo U.S. holidays occurred one day before, on, or one day after my birthday.")
```
